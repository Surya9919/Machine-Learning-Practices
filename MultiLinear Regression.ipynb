```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_excel("Multiple Regression/DS4.xlsx")
df['BIAS'] = 1
X=pd.DataFrame({"BIAS":df["BIAS"],"X1":df["X1"],"X2":df["X2"],"X3":df["X3"],"X4":df["X4"],"X5":df["X5"]})
print(X)
Y=np.array(np.random.rand(54,1))
Y
```

```python
XT=np.transpose(X)

XT
```

```python
XXT=XT.dot(X)
XXT
```

```python
XXTI=np.linalg.inv(XXT)
XXTI
```

```python
LHS=XT .dot(Y)
LHS
```

```python
ANS=XXTI.dot(LHS)
ANS
```

```python
print("A0",ANS[0])
print("A1",ANS[1])
print("A2",ANS[2])
print("A3",ANS[3])
print("A4",ANS[4])
print("A5",ANS[5])
X1max=np.max(df["X1"])+100
X2max=np.max(df["X2"])+100
X3max=np.max(df["X3"])+100
X4max=np.max(df["X4"])+100
X5max=np.max(df["X5"])+100
Xmin=np.min(Y)-100
x1=np.linspace(Xmin,X1max,100)
x2=np.linspace(Xmin,X2max,100)
x3=np.linspace(Xmin,X3max,100)
x4=np.linspace(Xmin,X4max,100)
x5=np.linspace(Xmin,X5max,100)
```

```python
Line=ANS[0]+ANS[1]*x1+ANS[2]*x2+ANS[3]*x3+ANS[4]*x4+ANS[5]*x5
```

```python
plt.plot(df.X1,Line,c='red',label="Regression line")
```

```python
plt.scatter(df["X1"],Y,label='X1')
plt.scatter(df["X2"],Y,label='X2')
plt.scatter(df["X3"],Y,label='X3')
plt.scatter(df["X4"],Y,label='X4')
plt.scatter(df["X5"],Y,label='X5')
plt.plot(x1,Line,c='red',label='Regression line')
plt.plot(x2,Line,c='blue',label='Regression line')
plt.plot(x3,Line,c='green',label='Regression line')
plt.plot(x4,Line,c='black',label='Regression line')
plt.plot(x4,Line,c='yellow',label='Regression line')
plt.legend()
plt.show()
```
