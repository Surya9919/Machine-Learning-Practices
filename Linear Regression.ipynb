```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
def LinearRegression(X,Y):
    df["x^2"] = X ** 2
    df["xy"] = X* Y
    ymean=Y.mean()
    xmean=X.mean()
    xymean=df['xy'].mean()
    xxmean=df['x^2'].mean()
    a1=(xymean-(xmean*ymean))/(xxmean-(xmean*xmean))
    a0=ymean-(a1*xmean)
    return a0,a1
```

```python
df=pd.read_excel("Linear Regression/DS2.xlsx")
xdata=df['X']
ydata=df['Y']
a0 ,a1 = LinearRegression(xdata,ydata)
print(a0)
print(a1)
```

```python
# plot points and line plot
Xmax=np.max(xdata)+100
Xmin=np.min(ydata)-100
x=np.linspace(Xmin,Xmax,100)
yline2=a0+(a1*df['X'])
yline=a0+(a1*x)
plt.plot(df['X'],yline2,c='red',label='Regression line')
plt.plot(x,yline,c='red',label='Regression line')
plt.scatter(xdata,ydata)
plt.show()
```
